

### Hierarchical Attention Networks for Document Classification    

#### 摘要

  本文提出文档分类的分层注意力网络。文本模型有两个显著不同特征：（1）模型有与文档分层结构相对应的分层结构。（2）模型使用注意力在字级别和句子级别，使得构成文档描述注意不同重要内容。实验在六个大规模文本分类，结果显示已提出架构超过之前方法。注意力层可视化显示模型选择定性信息字和句子。

##### 基于GRU序列编码

$$
h_t=(1-z_t) \odot h_{t-1} + z_t \odot \tilde{h_{t}}
$$

$$
z_t=\sigma(W_zx_t+U_zh_{t-1}+b_z)
$$

$$
\tilde{h_t}=tanh(W_hx_t+r_t\odot(U_hh_{t-1})+b_h)
$$

$$
r_t=\sigma(W_rx_t+U_rh_{t-1}+b_r)
$$

##### 分层注意力

​      假设一篇文档有L个句子$s_i$,每个句子包含$T_i$个单词，$\omega_{it}$表示第i个句子的第t个字。提出的模型映射原始文档为一个向量描述，之后进行文档分类。

***字编码***，通过GRU进行编码：


$$
x_{it}=W_ew_{it},t \in [1,T]
$$

$$
\vec{h_{it}}=\vec {GRU}(x_{it}),t \in [1,T]
$$

$$
\overleftarrow{h_{it}}=\overleftarrow{GRU}(x_{it}),t \in [T,1]
$$

$$
h_{it}=[\vec{h_{it}},\overleftarrow{h_it}]
$$

***字注意力***，并不是每个字对句子含义贡献一样，这里使用注意力进行处理：
$$
u_{it}=tanh(W_wh_{it}+b_\omega)
$$

$$
\alpha_{it}=\frac{exp({u_{it}^Tu_\omega)}}{\sum_texp({u_{it}^Tu_\omega)}}
$$

$$
s_i=\sum_t{\alpha_{it}h_{it}}
$$
其中$u_{\omega}$是个随机初始化并且需要训练的。

***句子编码***和***句子注意力***与字编码和字注意力类似，不再累述。
